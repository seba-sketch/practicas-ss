```python
class Libro:
    def __init__(self, titulo, autor, anio_publicacion):
        self.titulo = titulo
        self.autor = autor
        self.anio = anio_publicacion
        self.prestado = False # Atributo por defecto
    def prestar(self):
        if not self.prestado:
            self.prestado = True
            return f"El libro '{self.titulo}' ha sido prestado."
        else:
            return f"El libro '{self.titulo}' ya está prestado."
    def descripcion(self):
        return f"Título: {self.titulo}, Autor: {self.autor}, Año: {self.anio}"
libro1 = Libro("Cien años de soledad", "Gabo", 1967)
libro2 = Libro("Rayuela", "Cortázar", 1963)
print(libro1.descripcion()) 
print(libro2.descripcion())

print("-" * 20)
print(libro1.prestar())
print(libro1.prestar()) 
print(f"¿Libro 2 prestado? {libro2.prestado}")
```

```python
# listas 2D
# Una Lista 2D es simplemente una Lista donde cada elemento es, 
# a su vez, otra Lista. Esto te permite crear estructuras que se asemejan a tablas, 
# cuadrículas o matrices.
# 1. Creación de una Lista 2D (una matriz de 3x3)
matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
# 2. Acceso por doble índice: [Fila][Columna]
elemento_central = matriz[1][1] 
print(f"Elemento central: {elemento_central}")
# 3. Acceso al primer elemento de una Fila
print(f"Elemento: {matriz[2][0]}")
# 4. Modificar un elemento
matriz[2][2] = 100
print(f"Matriz completa después del cambio: {matriz}")
```

```python
# while
contador = 0
limite = 5
# El bucle se repite MIENTRAS la condición sea Verdadera
while contador < limite:
    
    print(f"Contando... {contador}")
    # ¡CRUCIAL! Debes modificar la variable de control (contador),
    # de lo contrario, el bucle se repetiría infinitamente.
    contador = contador + 1 
print("¡Bucle terminado!") 
```

```python
numero_secreto = 7
adivinado = False

# ✅ Define el intento FUERA para que el bucle lo use
intento = 7 # Le damos el valor correcto de 7 para que termine

while adivinado == False:
    # Si el intento fuera 6, se ejecuta el 'else', y se repite.
    
    if intento == numero_secreto:
        print("¡Ganaste!")
        adivinado = True # Esto hace que el bucle termine
    else:
        # Si el intento fuera 6, el programa se queda aquí repitiendo el mensaje.
        print("Sigue intentando") 
        break # ⬅️ Opcional: El 'break' también puede forzar la salida de un bucle

print("Fin del programa.")
```

```python
def division_segura():
    entrada_usuario = input("Ingresa un numero: ")
    try:
        numero = int(entrada_usuario)
        division = 10 / numero
        print(f"Resultado: {division}")
    except ValueError:
        print("Error: Debes ingresar un numero entero valido.")
    except ZeroDivisionError: 
            print("Error: No se puede dividir entre cero")
    print("--------------------")
division_segura()
division_segura()
division_segura()
```

```python
# proyecto: Diccionario
# 1. El Diccionario actúa como la "base de datos" de traducciones.
traducciones = {
    "lista": "List",
    "diccionario": "Dictionary",
    "clase": "Class",
    "bucle": "Loop",
    "si": "If",
    "casa": "house"
}
def traducir_palabra(palabra_espanol):
    palabra_espanol = palabra_espanol.lower()
    traduccion = traducciones.get(palabra_espanol, "¡Palabra no encontrada!")
    return traduccion
palabra1 = "Bucle"
resultado1 = traducir_palabra(palabra1)
print(f"'{palabra1}' se traduce como: {resultado1}")

palabra2 = "pala"
resultado2 = traducir_palabra(palabra2)
print(f"'{palabra2}' se traduce como: {resultado2}")
```

```python
traducciones = {
    "lista": "List",
    "diccionario": "Dictionary",
    "clase": "Class",
    "bucle": "Loop",
    "si": "If"
}
def traducir_seguro(palabra_espanol):
    if palabra_espanol == "":
        return "Error: Ingresa una palabra."
    palabra_espanol = palabra_espanol.lower()
    return traducciones.get(palabra_espanol, "Palabra no encontrada!") 
    
print(traducir_seguro("clase"))     # Exitoso, mayúscula
print(traducir_seguro("casa"))      # No existe
print(traducir_seguro(""))          # Cadena vacía
```
